```python
import pandas as pd
from lib.toml_utils import read_toml
```

```python
# get the metrics output
df = read_toml('../cache/calculate_metrics_output.toml', "contracts")
df
```

```python
# select issues column
if df["issues"] is not None:
    df['issues'] = df["issues"].apply(lambda issues: list(map(lambda i: i['category'], issues)) if issues is not None else None)

df
```

```python
# Get keys that have x in its name
def get_keys(df, x):
    return list(filter(lambda key: x in key, df.columns))
```

```python
recall_keys = get_keys(df, 'recall')
positive_keys = get_keys(df, 'positive')
negative_keys = get_keys(df, 'negative')


# filter the dataframe to only have the recall and prediction columns 
# and sort by recall and only select all recall columns < 0.5
df[recall_keys + positive_keys + negative_keys + ['issues']].where(df[recall_keys].max(axis=1) < 0.5)
```
